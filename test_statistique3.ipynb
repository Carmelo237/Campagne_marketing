```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les donnÃ©es
file_path = 'C:\\Users\\marye\\Downloads\\Business_Case (1).csv'  # Chemin vers votre fichier CSV
data = pd.read_csv(file_path, delimiter=';')

# Supprimer les doublons basÃ©s sur user_id
unique_users = data.drop_duplicates(subset='user_id')

# Compter le nombre d'ab_cohort_name pour chaque user_id unique
cohort_counts = unique_users['ab_cohort_name'].value_counts()

# Convertir en DataFrame pour une manipulation plus facile
cohort_counts_df = cohort_counts.reset_index()
cohort_counts_df.columns = ['ab_cohort_name', 'user_count']

# Afficher les rÃ©sultats
print(cohort_counts_df)
# Tracer l'histogramme
plt.figure(figsize=(12, 6))
sns.barplot(x='ab_cohort_name', y='user_count', data=cohort_counts_df, palette='viridis', hue='ab_cohort_name')
plt.title("FrÃ©quence des Compagnies Publicitaires (ab_cohort_name) par Utilisateur (user_id)")
plt.xlabel("Compagnie Publicitaire (ab_cohort_name)")
plt.ylabel("Nombre d'Utilisateurs (user_count)")
plt.xticks(rotation=45)
plt.show()

# Calculer et afficher la moyenne des utilisateurs par ab_cohort_name
mean_users = cohort_counts_df['user_count'].mean()
print(f"Moyenne des utilisateurs par ab_cohort_name : {mean_users}")                                       
```

```python
# VÃ©rifier les valeurs manquantes
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# AgrÃ©ger les donnÃ©es par cohorte A/B
cohort_stats = data.groupby('ab_cohort_name').agg({
    'session_length': ['mean', 'median', 'count'],
    'publisher_revenue': ['mean', 'median']
}).reset_index()

# Renommer les colonnes pour plus de clartÃ©
cohort_stats.columns = ['Cohorte', 'DurÃ©e moyenne', 'DurÃ©e mÃ©diane', 'Nombre de sessions', 'Revenu moyen', 'Revenu mÃ©dian']
print(cohort_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Graphique pour comparer les durÃ©es moyennes des sessions par cohorte
plt.figure(figsize=(10,6))
sns.barplot(x='Cohorte', y='DurÃ©e moyenne', data=cohort_stats, color='b')
plt.title('DurÃ©e moyenne des sessions par cohorte')
plt.show()

# Graphique pour comparer les revenus moyens par cohorte
plt.figure(figsize=(10,6))
sns.barplot(x='Cohorte', y='Revenu moyen', data=cohort_stats, color='g')
plt.title('Revenu moyen par cohorte')
plt.show()
```

```python
# Convertir les colonnes 'open_at' et 'acquired_at' en datetime si ce n'est pas dÃ©jÃ  fait
data['open_at'] = pd.to_datetime(data['open_at'])
data['acquired_at'] = pd.to_datetime(data['acquired_at'])

# CrÃ©er une colonne pour la date de la session (jour sans l'heure)
data['jour_session'] = data['open_at'].dt.date

# Compter le nombre de sessions par utilisateur et par jour
user_sessions_per_day = data.groupby(['user_id', 'jour_session'])['session_id'].count().reset_index()

# Joindre les cohortes publicitaires Ã  chaque utilisateur
user_sessions_per_day = user_sessions_per_day.merge(data[['user_id', 'ab_cohort_name']].drop_duplicates(), on='user_id')

# CrÃ©er une colonne 'abandon' : Si l'utilisateur n'a qu'une session ce jour-lÃ , il est considÃ©rÃ© comme "perdu"
user_sessions_per_day['abandon'] = user_sessions_per_day['session_id'].apply(lambda x: 1 if x == 1 else 0)

# Compter le nombre d'utilisateurs perdus par cohorte et par jour
abandon_par_jour = user_sessions_per_day.groupby(['jour_session', 'ab_cohort_name'])['abandon'].sum().reset_index()

# Renommer les colonnes pour plus de clartÃ©
abandon_par_jour.columns = ['Jour', 'FrÃ©quence de publicitÃ©', 'Nombre d\'utilisateurs perdus']
print(abandon_par_jour)
```

```python
# Visualiser le nombre d'utilisateurs perdus par jour en fonction de la frÃ©quence des publicitÃ©s
plt.figure(figsize=(12,6))
sns.lineplot(x='Jour', y='Nombre d\'utilisateurs perdus', hue='FrÃ©quence de publicitÃ©', data=abandon_par_jour)
plt.title('Nombre quotidien d\'utilisateurs perdus par frÃ©quence de publicitÃ©')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Nombre total d'utilisateurs initiaux
total_utilisateurs_initials = data['user_id'].nunique()

# 2. Nombre d'utilisateurs retenus
utilisateurs_retenus = data.groupby('user_id')['session_id'].count().apply(lambda x: x > 1).sum()

# 3. Calculer le nombre d'utilisateurs perdus
utilisateurs_perdus = total_utilisateurs_initials - utilisateurs_retenus

# 4. Affichage des rÃ©sultats
print("Nombre total d'utilisateurs initiaux :", total_utilisateurs_initials)
print("Nombre d'utilisateurs retenus :", utilisateurs_retenus)
print("Nombre d'utilisateurs perdus :", utilisateurs_perdus)
```

```python
# Supposons que le nombre total d'utilisateurs est la somme des utilisateurs retenus et perdus
utilisateurs_totaux = abandon_par_jour['Nombre d\'utilisateurs perdus'].sum()

# Boucle Ã  travers chaque frÃ©quence de publicitÃ© pour gÃ©nÃ©rer un graphique circulaire pour chaque catÃ©gorie
frequences = abandon_par_jour['FrÃ©quence de publicitÃ©'].unique()

for freq in frequences:
    # Filtrer les donnÃ©es pour chaque frÃ©quence de publicitÃ©
    data_freq = abandon_par_jour[abandon_par_jour['FrÃ©quence de publicitÃ©'] == freq]
    
    # Calculer le nombre total d'utilisateurs perdus pour cette frÃ©quence
    total_perdus = data_freq['Nombre d\'utilisateurs perdus'].sum()
    
    # Calculer le nombre d'utilisateurs retenus
    total_retenus = utilisateurs_totaux - total_perdus

    # DonnÃ©es pour le graphique circulaire
    labels = ['Utilisateurs retenus', 'Utilisateurs perdus']
    values = [total_retenus, total_perdus]

    # CrÃ©er le graphique circulaire
    plt.figure(figsize=(6, 6))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
    plt.title(f'Proportion d\'utilisateurs retenus vs. perdus pour {freq}')
    plt.axis('equal')  # Assurer que le graphique soit circulaire
    
    # Afficher le graphique
    plt.show()
```

```python
 import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# AgrÃ©ger les donnÃ©es par cohorte A/B
cohort_stats = data.groupby('ab_cohort_name').agg({
    'session_length': ['mean', 'median', 'count'],
    'publisher_revenue': ['mean', 'median']
}).reset_index()

# Renommer les colonnes pour plus de clartÃ©
cohort_stats.columns = ['Cohorte', 'DurÃ©e moyenne', 'DurÃ©e mÃ©diane', 'Nombre de sessions', 'Revenu moyen', 'Revenu mÃ©dian']

# Calcul des pourcentages pour la durÃ©e moyenne des sessions
total_session_length = cohort_stats['DurÃ©e moyenne'].sum()
cohort_stats['Pourcentage durÃ©e moyenne'] = (cohort_stats['DurÃ©e moyenne'] / total_session_length) * 100

# Calcul des pourcentages pour les revenus moyens
total_revenue = cohort_stats['Revenu moyen'].sum()
cohort_stats['Pourcentage revenu moyen'] = (cohort_stats['Revenu moyen'] / total_revenue) * 100

# Graphique circulaire pour les pourcentages de la durÃ©e moyenne des sessions
plt.figure(figsize=(8,8))
plt.pie(cohort_stats['Pourcentage durÃ©e moyenne'], labels=cohort_stats['Cohorte'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Blues"))
plt.title('RÃ©partition de la durÃ©e moyenne des sessions par cohorte (%)')
plt.show()

# Graphique circulaire pour les pourcentages des revenus moyens par cohorte
plt.figure(figsize=(8,8))
plt.pie(cohort_stats['Pourcentage revenu moyen'], labels=cohort_stats['Cohorte'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Greens"))
plt.title('RÃ©partition du revenu moyen par cohorte (%)')
plt.show()

# Afficher les donnÃ©es finales avec les pourcentages
print(cohort_stats)
```
