```python
import pandas as pd

# Charger les donnÃ©es
file_path = 'C:\\Users\\marye\\Downloads\\Business_Case (1).csv'  # Chemin vers votre fichier CSV
data = pd.read_csv(file_path, delimiter=';')

# Supprimer les doublons basÃ©s sur user_id
unique_users = data.drop_duplicates(subset='user_id')

# Compter le nombre d'ab_cohort_name pour chaque user_id unique
cohort_counts = unique_users['ab_cohort_name'].value_counts()

# Afficher les rÃ©sultats
print(cohort_counts)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les donnÃ©es
file_path = 'C:\\Users\\marye\\Downloads\\Business_Case (1).csv'  # Chemin vers votre fichier CSV
data = pd.read_csv(file_path, delimiter=';')

# Compter le nombre d'ab_cohort_name
cohort_counts = data['ab_cohort_name'].value_counts()

# CrÃ©er un DataFrame pour la visualisation
cohort_counts_df = pd.DataFrame(cohort_counts).reset_index()
cohort_counts_df.columns = ['ab_cohort_name', 'user_count']

# Tracer l'histogramme
plt.figure(figsize=(12, 6))
sns.barplot(x='ab_cohort_name', y='user_count', data=cohort_counts_df, palette='viridis', hue='ab_cohort_name')
plt.title("FrÃ©quence des Compagnies Publicitaires (ab_cohort_name) par Utilisateur (user_id)")
plt.xlabel("Compagnie Publicitaire (ab_cohort_name)")
plt.ylabel("Nombre d'Utilisateurs (user_count)")
plt.xticks(rotation=45)
plt.show()
```
